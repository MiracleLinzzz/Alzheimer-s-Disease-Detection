```python
import pandas as pd
import seaborn as sns
import ast
import matplotlib.pyplot as plt
import numpy as np
```

```python
mean = pd.read_csv("/content/gdrive/MyDrive/mean_val_acc.csv")
last = pd.read_csv("/content/gdrive/MyDrive/last_val_acc.csv")
test = pd.read_csv("/content/gdrive/MyDrive/test_acc.csv")
```

```python
mean = mean.to_dict(orient='index')[0]
for key, value in mean.items():
  mean[key] = ast.literal_eval(value)
```

```python
df = pd.DataFrame(mean)
df = df.stack().unstack(0)
df = df.stack().reset_index().drop(columns="level_1").rename(columns={"level_0":"seed", 0:"accuracy"})
```

```python
sns.set(font_scale=1,style="whitegrid",rc={'figure.figsize':(8,6), 'figure.dpi':100})
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.set_title("Results of different seeds without cross validation")
ax = sns.swarmplot(x="seed",y="accuracy",data=df,size=6,linewidth=1)
```

```python
sns.set(font_scale=1,style="whitegrid",rc={'figure.dpi':100})
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.set_title("Mean validation accuracy of different seeds without cross validation")
ax = sns.boxplot(x="seed",y="accuracy",data=df,whis=np.inf)
ax = sns.swarmplot(x="seed",y="accuracy",data=df,color=".2")
```
